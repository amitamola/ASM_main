### Creating a custom VIF that uses adjusted R_square instead of usual R_square

```python
from statsmodels.regression.linear_model import OLS

def vif_adjusted(X, exog_idx):
    """
    Compute the variance inflation factor (VIF) for a given feature in a dataset, using the adjusted R squared.
    
    Parameters
    ----------
    X : array_like
        The design matrix, with shape (n_samples, n_features).
    exog_idx : int
        The index of the feature for which to compute the VIF.
        
    Returns
    -------
    vif : float
        The variance inflation factor for the given feature.
    """
    # Extract the feature for which to compute the VIF
    exog = X[:, exog_idx]
    
    # Compute the OLS regression model for the feature against all other features, including an intercept term
    ols_model = OLS(endog=exog, exog=X[:, np.arange(X.shape[1]) != exog_idx], hasconst=True)
    ols_result = ols_model.fit()
    
    # Compute the VIF using the adjusted R squared value
    vif = 1 / (1 - ols_result.rsquared_adj)
    
    return vif
```

```python
for val in range(10, 121, 10):
    print(f'Adj VIF for column 3 with {val}  columns is: {vif_adjusted(X.values[:,0:val], 3)}')
```

```python
for val in range(10, 121, 10):
    print(f'Adj VIF for column 7 with {val}  columns is: {vif_adjusted(X.values[:,0:val], 7)}')
```
